


## <span style="color:green">Normal Distribution</span>

The probability density function of the normal distribution, first derived by De Moivre and 200 years later by both Gauss and Laplace independently, is often called the bell curve because of its characteristic shape.

The normal distributions occurs often in nature. For example, it describes the commonly occurring distribution of samples influenced by a large number of tiny, random disturbances, each with its own unique distribution.

```python
x = np.arange(0.0, 10.0, 0.1)
y = 3.0 * x + 1.0

noise = np.random.normal(0.0, 1.0, len(x))

plt.plot(x, y+noise, "r.", label="Actual")
plt.plot(x, y, "b-", label="Model")

plt.title("Simple Plot")
plt.ylabel("Mass")
plt.xlabel("Weight")
plt.legend()

plt.show()
```

Normal distribution is often used in studies and social experiements as it depicts an accurate real world expectation of random results. where the random values gravitate towards the central values and are less frequent as you reach the upper and lower limits.

## <span style="color:green">Exponential Distribution</span>

Exponential functions are used when you are trying to determine or estimate the time between two events occuring, similarly link to poisson distribution which is used to estimate the amount of events occur in X amount of time.

```python
target = 250
beta = 1.0/target

Y = np.random.exponential(beta, 5000)
plt.hist(Y, density=True, bins=200,lw=0,alpha=.8)
plt.plot([0,max(Y)],[target,target],'r--')
plt.ylim(0,target*1.1)
plt.show()
```

### Wikipedia Explaination
In probability theory and statistics, the exponential distribution (also known as the negative exponential distribution) is the probability distribution that describes the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate. It is a particular case of the gamma distribution. It is the continuous analogue of the geometric distribution, and it has the key property of being memoryless. In addition to being used for the analysis of Poisson point processes it is found in various other contexts.

The exponential distribution is not the same as the class of exponential families of distributions, which is a large class of probability distributions that includes the exponential distribution as one of its members, but also includes the normal distribution, binomial distribution, gamma distribution, Poisson, and many others.

## <span style="color:green">Dirichlet Distribution</span>

```python
s = np.random.dirichlet((10, 5, 3), 20).transpose()

plt.barh(range(20), s[0])
plt.barh(range(20), s[1], left=s[0], color='g')
plt.barh(range(20), s[2], left=s[0]+s[1], color='r')
plt.title("Lengths of Strings")
```

## <span style="color:green">Gamma Distribution</span>

<span style="color:black">The Gamma distribution is widely used in engineering, science, and business, to model continuous variables that are always positive and have skewed distributions. In Swedge, the Gamma distribution can be useful for any variable which is always positive, such as cohesion or shear strength for example.</span>

```python
import scipy.stats as stats 

a = 0.5
b = 0.5

x = np.arange(0.01, 1, 0.01)
y = stats.gamma.pdf(x, a=29, scale=0.33333)

plt.plot(x, y)
plt.title('Gamma Distribution')
plt.xlabel('x')
plt.ylabel('Probability')
plt.show()
```


## <span style="color:green">Beta Distribution</span>

<span style="color:black">The Beta distribution is a special case of the Dirichlet distribution, and is related to the Gamma distribution. It has the probability distribution function. It is often seen in Bayesian inference and order statistics.</span>

```python
from scipy.stats import beta

a = 0.5
b = 0.5

x = np.arange(0.01, 1, 0.01)
y = beta.pdf(x, a, b)

plt.plot(x, y)
plt.title('Beta Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

## <span style="color:green">Binomial Distribution</span>

<span style="color:black">A frequency distribution of the possible number of successful outcomes in a given number of trials in each of which there is the same probability of success.</span>

```python
from scipy.stats import binom

fig, ax = plt.subplots(1, 1)

x = range(6)
n, p = 5, 0.166

rv = binom(n, p)

ax.vlines(x, 0, rv.pmf(x), colors='b', linestyles='-', lw=1 , label='Probablity')           

plt.show()
```

## <span style="color:green">Poisson Distribution</span>

The posson distribution is used mainly for determining things like the number of events that occur in a specific amount or time.

```python
poisson = np.random.poisson(0.5, 1000)

plt.hist(poisson)
plt.show()
```

```python
normal = np.random.normal(0.5, 2*np.sqrt(2*np.log(2)), 1000)
poisson = np.random.poisson(0.5, 1000)

plt.hist(normal)
plt.hist(poisson)
plt.show()
```
